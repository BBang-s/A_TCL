    탑승이력에 학교, 공항,항공, 터미널, 아파트 등을 category화 해보자.

```python
def category(station_name):
    if '항공' in station_name :
        return 1
    elif '공항' in station_name :
        return 1
    elif '학교' in station_name :
        return 2
    elif '터미널' in station_name :
        return 3
    elif '아파트' in station_name :
        return 4
    else : return 5
```

```python
bts['geton_station_name'] = bts['geton_station_name'].astype(str, inplace = True)
bts['getoff_station_name'] = bts['getoff_station_name'].astype(str, inplace = True)
```

```python
bts['geton_station_catg'] = bts['geton_station_name'].apply(lambda x : category(x))
bts['getoff_station_catg'] = bts['getoff_station_name'].apply(lambda x : category(x))
```
